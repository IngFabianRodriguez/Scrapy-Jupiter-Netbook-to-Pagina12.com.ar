```python
def links(s_seccion_noticias):
    links_list = []
    for noticia in s_seccion_noticias:
        link = noticia.a.get('href')
        links_list.append(link)
    return links_list
```

```python
noticias_dia = links(s_seccion_noticias)
noticias_dia
```

```python
noticia = noticias_dia[0]
nota
```

```python
def scrapnote(noticia):
    list_element = {}
    try:
        nota = requests.get(noticia)
        if nota.status_code == 200:
            s_nota = BeautifulSoup(nota.text, 'lxml')
            # Extraemos titulo
            title = s_nota.find('h1')
            title = title.text
            list_element['titulo'] = title
            # Extraemos fecha
            fecha = s_nota.find('span', attrs={'pubdate':'pubdate'}).get('datetime')
            list_element['fecha'] = fecha
            # Extraemos prefijo
            prefijo = s_nota.find('h2', attrs={'class': 'article-prefix'})
            prefijo = prefijo.text
            list_element['prefijo'] = prefijo
            # Extraemos los parrafos
            parrafos = s_nota.find('div', attrs={'class': 'article-text'}).find_all('p')
            # iteramos en elementos parrafos
            for parrafo in parrafos:
                parrafo = parrafo.text
                list_element['parrafo'] = parrafo
            # Extraemos las imagenes
            media = s_nota.find('div', attrs={'class': 'article-main-media-image'}).find_all('img')
            for imagen in media:
                imagen = imagen.get('data-src')
                list_element['imagen'] = imagen
        else:
            print('error')
    except Exception as e:
            print('Error')
            print(e)
            print('\n')
    return list_element
        
```

```python
noticias_dia
```

```python
noticia_dic = []
for noticia in noticias_dia:
    noticia_dic.append(scrapnote(noticia))
```
